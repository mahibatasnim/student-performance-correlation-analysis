```python
import pandas as pd
import zipfile
import requests
from io import BytesIO
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00320/student.zip"
response = requests.get(url)
zip_file=zipfile.ZipFile(BytesIO(response.content))
df = pd.read_csv(zip_file.open("student-mat.csv"), sep=';')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["failures"], df["G3"])
plt.xlabel("Past Failures")
plt.ylabel("Final Grade (G3)")
plt.title("Failures vs Final Grade")
plt.show()
```

**Correlation and Statistical Analysis of Factors Influencing Student Academic Performance**

To quantify and examine the relationship between selected academic (G1,G2) and behavioral variables(studytime,failures,absences, ECA) and student's final grades (G3) through correlation analysis and statistical visualization.

**Methodology**


*   Utilized a publicly available educational dataset from the UCI Machine Learning Repository.
*   Successfully able to conduct a structural and descriptive data inspection to understand variable types and distributions.


*   By assessing missing values verified dataset completeness.
*   To identify statistically relevant relationships performed correlation analysis on numeric variables.

*   Visualized key variable interactions using scatter plots to support quantitative findings.


**Key Findings**


*   Failures exhibit a moderate negative correlation with final grade (r ≈ -0.36).

*  All the study time shows a weak positive correlation with final grade  (r ≈ 0.09).

*   Absences shows almost no correlation with final grade (r ≈ 0.03).


**Conclusion**

The project analysis clearly states that among the variables, academic failures have the most substantial negativeimpact on students final grades. Which syggests that previously done academic difficulties strongly  predict lower performance. Study time appears to have minor positive influence which points that efforts are useful to get better grades. Shockingly, absences exhibits almost no correlation with final grades. So, only attendance us not a reliable predictor of performance. All these findings provide clear insight into which factors affect students the most and which almost does not affect at all in student achivement and not to mention it also can guide furthur research interventios.
